# Modelo predictivo  
### Separacion de datos

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_bal, y_bal, test_size=0.2, random_state=42)
```

### Creacion de modelos  
Modelo predictivo 1: Regresion logistica

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

modelo1 = LogisticRegression()

#Normalizamos los datos
scaler = StandardScaler()
X_train_scaler = scaler.fit_transform(X_train)
X_test_scaler = scaler.transform(X_test)
```

```python
#entrenamos el modelo 1
modelo1.fit(X_train_scaler, y_train)

#ahora hacemos las predicciones
y_pred1 = modelo1.predict(X_test_scaler)
```

Modelo predictivo 2: Arbol de decision

```python
from sklearn.tree import DecisionTreeClassifier

#no necesitamos normalizar los datos pues este modelo predictivo no es sensible a los outliers
modelo2 = DecisionTreeClassifier(max_depth=5, random_state=42)
modelo2.fit(X_train, y_train)
y_pred2 = modelo2.predict(X_test)
```

### Evaluacion de los modelos

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

#evaluacion del primer modelo
print("Evaluacion del primer modelo: Regresion Logistica\n\n", classification_report(y_test, y_pred1))
print("Evaluacion del segundo modelo: Arbol de decision\n\n", classification_report(y_test, y_pred2))
```

```python
## Matriz de confusion del primer modelo
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred1)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Purples", cbar=False,
            xticklabels=["No Canceló", "Canceló"],
            yticklabels=["No Canceló", "Canceló"])

plt.xlabel("Predicción")
plt.ylabel("Real")
plt.title("Matriz de Confusión del modelo 1: Regresion Logistica")
plt.show()
```

```python
#Matriz de confusion del segundo modelo
cm = confusion_matrix(y_test, y_pred2)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Purples", cbar=False,
            xticklabels=["No Canceló", "Canceló"],
            yticklabels=["No Canceló", "Canceló"])

plt.xlabel("Predicción")
plt.ylabel("Real")
plt.title("Matriz de Confusión del modelo 2: Arbol de decisiones")
plt.show()
```

El mejor modelo es el de regresion logistica ya que tiene mejor reporte que el arbol de decisiones, tiene menos errores. Detecta una gran cantidad de personas que cancelaron el servicio. No muestra overfitting.


# Interpretacion y conclusiones  
### Analisis de la importancia de las variables

```python
import pandas as pd
import numpy as np


coef = modelo1.coef_[0]

# DataFrame con importancia en %
feature_importance = pd.DataFrame({
    "Variable": X_train.columns,
    "Importancia":(np.abs(coef) / np.sum(np.abs(coef)))
}).sort_values(by="Importancia", ascending=False)

print(feature_importance)

```

```python
importances = modelo2.feature_importances_

# Crear DataFrame
feature_importance_tree = pd.DataFrame({
    "Variable": X_train.columns,
    "Importancia": importances
}).sort_values(by="Importancia", ascending=False)

print(feature_importance_tree)
```

Para el modelo de regresion logistica hay mas variables importantes


### Conclusion  
El mejor modelo que se visualiza para el estudio es Regresión Logística ya que ofrece menos errores y mejor equilibrio entre precisión y recall que el Árbol de Decisión. La Regresión Logística supera al Árbol en todas las métricas clave. El Árbol muestra más Falsos Negativos (pierde más clientes que realmente cancelan), señal de menor generalización.   

Los principales factores de cancelación son:  
- Contrato mensual: mayor probabilidad de evasion.  
- Precio: se encuentran muchos outliers de gente que paga mucho y cancela el contrato.
- Servicios: hay mas cancelaciones de clientes que no tienen servicio de seguridad.  
- Fibra optica: los que tienen este servicio suelen cancelar.  

Algunas recomendaciones podrian ser promociones si es que se pasa a un mayor tiempo de contrato, promociones para clientes que tienen muchos servicios y estan pagando altos valores, agregar paquetes de seguridad. Incentivo del pago con debito automatico.
